## Информация фишера

В этой заметке будет введено понятие информации фишера.

Предположим состояние системы задается неизвестным параметром $a$. Мы наблюдаем некоторые значения $y_1, y_2, ..., y_N$ из системы с целью оценить $a$. Мы предполагаем, что данные содержат необходимую информацию о параметре $a$. Эти данные соотносятся со значением нашего параметра $a$ через распределение $p(y \rvert x)$. Основной задачей является измерить количество информации о параметре $a$ в данных $y$, определить эту информацию через свойства системы.

Итак, мы хотим оценить параметр $a$. Предположим существует некоторая функция $\hat{a}(y)$ оценки параметра $a$ по данным $y$. Также предположим, что получаемые ошибки несмещенные, тогда:

$$\int (\hat{a}(y) - a) p(y \rvert a) \,dy = 0$$

Продифференцируя это уравнение по $\partial / \partial a$, получим:

$$\int \,dy (\hat{a}(y) - a) \frac{\partial p}{\partial a} - \int \,dy p = 0$$

С учетом того, что $\partial p / \partial a = p \partial ln(p) / \partial a$ и \int \,dy p = 1, получаем:

$$\int \,dy (\hat{a}(y) - a) \frac{\partial ln(p)}{\partial a} p = 1$$

$$1 = \int \,dy \left[(\hat{a}(y) - a) \sqrt{p} \right] \left[\frac{\partial ln(p)}{\partial a} \sqrt{p} \right]$$

Далее используем неравенство [Коши — Буняковского](https://ru.wikipedia.org/wiki/Неравенство_Коши_—_Буняковского), получаем:

$$1 \leq \int \,dy \left[(\hat{a}(y) - a) \sqrt{p} \right]^2 \int \left[\frac{\partial ln(p)}{\partial a} \sqrt{p} \right]^2$$

$$1 \leq \int \,dy \left[(\hat{a}(y) - a)^2 p \right] \int \left[\left(\frac{\partial ln(p)}{\partial a}\right)^2 p \right]$$

Заметим, что оба полученных интеграла явлются математическим ожиданием некоторых велечин. Если посмотреть внимательно, то видно, что первый интеграл является средней ошибкой определения параметра $a$:

$$\int \,dy \left[(\hat{a}(y) - a)^2 p \right] = e^2$$

Второй интеграл и есть информация Фишера:

$$I(a) = \int \,dy \left[\left(\frac{\partial ln(p(y \rvert a))}{\partial a}\right)^2 p(y \rvert a) \right]$$

Таким образом средняя ошибка тесно связана с информацией Фишера:

$$e^2 = \frac{1}{I}$$
